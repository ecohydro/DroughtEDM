```python
# Import libraries
import os
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.lines as mlines
from matplotlib.colors import ListedColormap
import geopandas as gpd

# Load the box module from shapely to create box objects
from shapely.geometry import box
#import earthpy as et
#import seaborn as sns

# Ignore warning about missing/empty geometries
import warnings
warnings.filterwarnings('ignore', 'GeoSeries.notna', UserWarning)

# Adjust plot font sizes
#sns.set(font_scale=1.5)
#sns.set_style("white")

```

```python
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
```

```python
#attribute join
country_shapes = world[['geometry', 'iso_a3']]
country_names = world[['name', 'iso_a3']]
```

```python
#spatial join 
countries = world[['geometry', 'name']]
countries = countries.rename(columns={'name':'country'})
```

```python
country_shapes = country_shapes.merge(country_names, on='iso_a3')
country_shapes
```

```python
east_africa_countries = ('Burundi', 'Djibouti', 'Eritrea', 'Ethiopia', 'Kenya', 'Rwanda', 'Sudan', 'Somalia', 'South Sudan', 'Tanzania', 'Uganda')
```

```python
east_africa_shapes = country_shapes[country_shapes.name.isin(east_africa_countries)]
east_africa_shapes
```

```python
gadm_eastafrica = '/home/rgreen/tana-spin/rgreen/DroughtEDM/Data/AdminBoundaries/gadm36_EastAfrica.shp'
ecoregions_globe = '/home/rgreen/tana-spin/rgreen/DroughtEDM/Data/Ecoregions/Ecoregions2017.shp'
```

```python
ea_adminbndry = gpd.read_file(gadm_eastafrica)
ecoregions = gpd.read_file(ecoregions_globe)
```

```python
afrotropic = ecoregions[ecoregions.REALM == 'Afrotropic']
```

```python
# Plot the data
fig, ax = plt.subplots(figsize=(12, 8))

ea_adminbndry.plot(alpha=.5,
                         ax=ax)

afrotropic.plot(cmap='Greys',
                       ax=ax,
                       alpha=.5)

plt.axis('equal')
ax.set_axis_off()
plt.show()
```
