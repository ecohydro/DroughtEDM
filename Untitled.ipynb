# Code to shuffle anomalies and conduct EDM analysis


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def anomaly(data):
    """ Returns a set of anomalies, which are the diff between the value and mean.
    
    """
    return data - data.mean()

def norm(data):
    """ Normalizes data to zero mean and unit variance 
    
    """
    return (data - np.mean(data)) / np.std(data)

```

```python
df = pd.read_csv('low_high_ts_df.csv')

df['dekad'] = np.mod(df.index.values,37)
dekads = df.groupby('dekad')
dekad_anom = dekads.transform(lambda x: (x - x.mean())) # Transform data into anomalies.
dekad_mean = dekads.transform(lambda x:  x.mean())

df_actual = df.apply(lambda x: (x - x.mean())/x.std())

df_surrogate = (dekad_mean + dekad_anom.sample(frac = 1).reset_index(drop=True)).apply(lambda x: (x - x.mean())/x.std())

# df_surrogate['NDVI_high_skill'].plot()
# df_actual['NDVI_high_skill'].plot()

plt.plot(df_actual['NDVI_high_skill'], df_surrogate['NDVI_high_skill'], '.')
plt.xlabel('Actual')
plt.ylabel('Surrogate')
```
